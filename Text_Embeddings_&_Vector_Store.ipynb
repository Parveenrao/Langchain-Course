```python
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
document = TextLoader('/content/Sample.txt').load()
print(document)
```

```python
spillter = CharacterTextSplitter(chunk_size = 1000, chunk_overlap = 0)
documents = spillter.split_documents(document)
```

```python
print(documents[0])
```

```python
from langchain_chroma import Chroma
embedding_model = GoogleGenerativeAIEmbeddings(model="models/embedding-001")
db = Chroma.from_documents(documents,GoogleGenerativeAIEmbeddings(model = "models/embedding-001"))
```

```python
query = 'The point of using Loreum Ipsum'
docs = db.similarity_search(query)
print(docs[0].page_content)
```

**Search By Vector**

```python
ebmeed_vector = embedding_model.embed_query(query)
docs = db.similarity_search_by_vector(ebmeed_vector)
print(docs[0].page_content)
```
